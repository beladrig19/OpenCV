# Filtrage d'image

* Blur
* Gaussian blur
* Median blur

```python
# charger et afficher l'image "bois"

image_bois = imageio.imread('bois.png')

# afficher image

#cv2.imshow('Image Bois', image_bois)
#cv2.waitKey(0)
```

```python
# Utiliser la méthode "cv2.filter2D" avec un kernel de taille 2

# Creating the kernel(2d convolution matrix) 
kernel1 = np.ones((5, 5), np.float32)/30
  
# Applying the filter2D() function 
img_filter2d = cv2.filter2D(src=image_bois, ddepth=-1, kernel=kernel1)
```

```python
# Appliquer blur avec un kernel de taille 5x5
blur = cv2.blur(image_bois,(5,5))
  
plt.subplot(121),plt.imshow(image_bois),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# gaussian blur

# Appliquer blur avec un kernel de taille 5x5
gaussian_blur = cv2.GaussianBlur(image_bois,(5,5),0)
  
plt.subplot(121),plt.imshow(image_bois),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(gaussian_blur),plt.title('Gaussian_Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# median blur

image_train = imageio.imread('train.jpeg')

# Appliquer blur avec un kernel de taille 5x5
median_blur = cv2.medianBlur(image_train,5)
  
plt.subplot(121),plt.imshow(image_train),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median_blur),plt.title('Median_Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# bilateral filtering
img_bilateral = cv2.bilateralFilter(image_bois,10,75,75) 
# Displaying the output image
plt.imshow(img_bilateral)
```

# Détection des contours

```python
# charger l'image 02
# faites une recherche sur internet pour détecter les contours des objets dans image_02

image2 = imageio.imread('image_02.png')

edge_canny2 = cv2.Canny(image=image2, threshold1=100, threshold2=200)

# Display the edge image
plt.imshow(edge_canny2, cmap='gray')
```

#  Compter des objets

```python
# Compter les objets

min_area = 100  

contours, _ = cv2.findContours(edge_canny2, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

valid_contours = []
for contour in contours:
    area = cv2.contourArea(contour)
    if area > min_area:
        valid_contours.append(contour)


num_objects = len(valid_contours)

print(f"Nombre d'objets détectés : {num_objects}")
```

# Challenge
Compter le nombre d'objets dans l'image 03 avec le code précédent.
- Qu'est ce que vous remarquez?
- Utiliser l'algorithme Watershed pour detecter et séparer les objets connectés.

```python
# Lire l'image
image3 = cv2.imread('image_03.png')

# Convertir l'image en niveaux de gris
gray_image1 = cv2.cvtColor(image3, cv2.COLOR_BGR2GRAY)

# Appliquer le seuillage Otsu
ret3, otsu1 = cv2.threshold(gray_image1, 0, 255, cv2.THRESH_OTSU)

# Réduction du bruit
kernel = np.ones((3, 3), np.uint8)
opening = cv2.morphologyEx(otsu1, cv2.MORPH_OPEN, kernel, iterations=2)

# Sure background area
sure_bg = cv2.dilate(opening, kernel, iterations=3)

# Finding sure foreground area
dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
ret, sure_fg = cv2.threshold(dist_transform, 0.7 * dist_transform.max(), 255, 0)

# Finding unknown region
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg, sure_fg)

# Marker labelling
ret, markers = cv2.connectedComponents(sure_fg)

# Add one to all labels so that sure background is not 0, but 1
markers = markers + 1

# Now, mark the region of unknown with zero
markers[unknown == 255] = 0

# Convertir les marqueurs en CV_32SC1
markers = markers.astype(np.int32)

# Appliquer Watershed
watershed_result = cv2.watershed(image3, markers)

# Créer une copie de l'image d'origine pour afficher les marqueurs
output_image = image3.copy()
output_image[markers == -1] = [255, 0, 0]

# Afficher l'image résultante
plt.imshow(output_image)
plt.show()

```
